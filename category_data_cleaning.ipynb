```python
import csv
import datetime
import numpy as np
```

```python
with open('data/detroit.csv') as f:
    csv_file = csv.reader(f)
    next(csv_file)
    crime_list = []
    for row in csv_file:
        crime_list.append(row)
```

```python
data = []
for row in crime_list:
    crime_date = datetime.datetime.strptime(row[2], '%m/%d/%Y').date()
    crime_pos  = np.array([float(row[-2]), float(row[-1])])
    crime_type = row[3].split('-')[0].strip()
    data.append([crime_date, crime_pos, crime_type])
```

```python
data[0]
```

## Filter the one with '('

```python
for index, d in enumerate(data):
    data[index][2] = d[2].split('(')[0].strip()
    data[index][2] = d[2].split(',')[0].strip()
```

## Clean 'MURDER'

```python
data[3][-1]
```

```python
for index, d in enumerate(data):
    if d[-1][0] == 'MURDER':
        data[index][-1] = 'MURDER'
```

```python
def print_without_something(something):
    result = []
    for d in data:
        if d[-1] != something:
            result.append(d)
    return result


def print_with_something(something):
    result = []
    for d in data:
        if d[-1][0] == something:
            result.append(d)
    return result
```

## Clean 'HOMICIDE-FATAL SHOOTING' and 'HOMICIDE'

```python
for index, d in enumerate(data):
    if d[-1][0] == 'HOMICIDE' or d[-1][0] == 'HOMICIDE-FATAL SHOOTING':
        data[index][-1] = 'HOMICIDE'
```
